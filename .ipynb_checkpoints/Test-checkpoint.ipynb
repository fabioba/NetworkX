```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import networkx.algorithms.clique as clq
```

```python
G= nx.path_graph(8)
```

```python
#draw randomized graph path
nx.draw(G)
plt.show()
```

```python
#create new Graph
G=nx.Graph()
```

```python
#add nodes
G.add_node(1)
G.add_node(2)
G.add_node(3)
```

```python
#create edge between a pair of nodes
G.add_edge(1,2)
```

```python
nx.draw(G,with_labels=True)
plt.show()
```

```python
G.add_edge(1,3)
G.add_edge(2,3)
```

```python
#create a complete graph
nx.draw(G,with_labels=True)
plt.show()
```

```python
G.add_node(4)
G.add_edge(1,4)
```

```python
fig=plt.figure(figsize=(8,8))
nx.draw(G,with_labels=True)
plt.show()
```

```python
#return the maximal clique for each node
[item for item in clq.find_cliques(G)]
```
